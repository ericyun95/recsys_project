```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
# 컨텐츠의 특성(장르)에 관한 유사도 구하기 위해 나머지를 제거
df_vector = df.drop(columns = ['movie_id', 'movie_title', 'release_date'])
```

```python
df_vector.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_cosine = pd.DataFrame(cosine_similarity(df_vector, df_vector))
```

```python
df_cosine.head()
```

```python
a = df_cosine[0].sort_values(ascending = False)
```

```python
a
```

```python
def content_based(movie_id):
  # 코사인 유사도로 총 6개(검색하는 영화 제거용)
  recommend_list = df_cosine[movie_id-1].sort_values(ascending = False)[:6]
  recommend_list = recommend_list.drop(axis = 0, index = movie_id-1)
  # 유사 영화 인덱스와 유사도 값 저장
  recommend_index = list(recommend_list.index)
  similarity_values = recommend_list.values
  # 유사 영화 이름과 id 가져오기
  movie_list = df[['movie_id', 'movie_title']].loc[recommend_index]
  # 유사도 첨부
  movie_list['similarity'] = similarity_values
  return movie_list
```

```python
content_based(1)
```
